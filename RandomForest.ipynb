```python
import pandas as pd

# 파일 경로 지정
apply_train_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/apply_train.csv'
company_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/company.csv'
recruitment_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/recruitment.csv'
resume_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/resume.csv'
resume_certificate_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/resume_certificate.csv'
resume_education_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/resume_education.csv'
resume_language_path = 'C:/Users/82106/Desktop/국민대 이력서 공모전/open/resume_language.csv'

# 각 파일을 데이터프레임으로 불러오기
apply_train_df = pd.read_csv(apply_train_path)
company_df = pd.read_csv(company_path)
recruitment_df = pd.read_csv(recruitment_path)
resume_df = pd.read_csv(resume_path)
resume_certificate_df = pd.read_csv(resume_certificate_path)
resume_education_df = pd.read_csv(resume_education_path)
resume_language_df = pd.read_csv(resume_language_path)

# 각 데이터프레임의 처음 5행 출력
(apply_train_df.head(), 
 company_df.head(), 
 recruitment_df.head(), 
 resume_df.head(), 
 resume_certificate_df.head(), 
 resume_education_df.head(), 
 resume_language_df.head())
```

### 데이터 병합

```python
#필요한 특성을 선택하여 데이터 결합, 변환하는 과정

#이력서와 채용 공고 매칭 데이터에 채용 공고 관련 정보를 결합
apply_train_with_recruitment=pd.merge(apply_train_df,recruitment_df,on='recruitment_seq',how='left')
```

```python
#구직자의 이력서 정보를 결합
apply_train_with_resume=pd.merge(apply_train_with_recruitment,
                                resume_df,
                                on='resume_seq',
                                how='left')
#구직자의 자격증 정보를 결합
apply_train_with_certificate=pd.merge(apply_train_with_resume,
                                     resume_certificate_df,
                                     on='resume_seq',
                                     how='left')
#구직자의 교육 이력 정보를 결합
apply_train_with_education=pd.merge(apply_train_with_certificate,
                                   resume_education_df,
                                   on='resume_seq',
                                   how='left')
#구직자의 언어 능력 정보를 결합
final_data=pd.merge(apply_train_with_education,
                   resume_language_df,
                   on='resume_seq',
                   how='left')
final_data.head()
```

```python
#데이터의 기술적 통계량 및 상관관계 분석

#기술적 통계량
descriptive_stats=final_data.describe()

descriptive_stats
```

```python
#상관관계 분석
correlation_matrix=final_data.corr()

correlation_matrix
```

```python
#특성 엔지니어링 수행
#중요한 특성 선택 및 새로운 특성 생성

#특성 선택: 이력서 관련 주요 특성과 채용 공고 관련 주요 특성 선택
selected_feature=final_data[['resume_seq','recruitment_seq',
                              'education',
                             'degree','hope_salary',
                             'major_task','qualifications',
                             'career_month','univ_score',
                             'language','score']]
#새로운 특성 생성
#학력과 채용 공고의 학력 요구 사이의 일치도
#예시
selected_feature['education_match']=selected_feature.apply(lambda x: 1 if x['education']==x['degree']else 0,axis=1)

#구직자가 원하는 연봉과 채용 공고의 자격에 대한 연봉 기대치 비교
```

```python
#결측치 처리
#언어점수가 없는 경우 0으로 처리
selected_feature['score'].fillna(0,inplace=True)

#결과확인
selected_feature.info()
```

```python
X=selected_feature.drop(['recruitment_seq'],axis=1)
y=selected_feature['recruitment_seq']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report 
from sklearn.model_selection import cross_val_score

#결측치 처리: Nan 값을 적절한 값으로 대체

#결측치가 있는 특성 확인
nan_columns=X_train.columns[X_train.isna().any()].tolist()
```

```python
nan_columns
```

```python
for column in nan_columns:
    if selected_feature[column].dtype == 'object':
        #범주형 변수: 최빈값으로 대체
        mode_value= X_train[column].mode()[0]
        X_train[column].fillna(mode_value,inplace=True)
        X_test[column].fillna(mode_value,inplace=True)
        
    else:
        #연속형변수 : 평균으로 대체
        mean_value=X_train[column].mean()
        X_train[column].fillna(mean_value,inplace=True)
        X_test[column].fillna(mean_value,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# 문자열 열을 숫자로 변환하기 위한 LabelEncoder 생성
label_encoder = LabelEncoder()

# 문자열 열에 해당하는 열을 선택하여 변환
X_train['resume_seq'] = label_encoder.fit_transform(X_train['resume_seq'])
```

```python
from sklearn.preprocessing import LabelEncoder

# 문자열 열을 숫자로 변환하기 위한 LabelEncoder 생성
label_encoder = LabelEncoder()

# 문자열 열에 해당하는 열을 선택하여 변환
X_test['resume_seq'] = label_encoder.fit_transform(X_test['resume_seq'])
```

```python
rf_model = RandomForestClassifier(n_estimators=10, 
                                  max_depth=5,
                                      random_state=42)
#모델 학습 및 성능 평가
rf_model.fit(X_train,y_train)
```

```python
cv_scores=cross_val_score(rf_model,X_train,y_train,cv=5)
```

```python
#테스트 데이터를 사용한 성능 평가
y_pred=rf_model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
classification_rep=classification_report(y_test,y_pred)

cv_scores
```

```python
#전체 이력서와 채용 공고의 조합 생성 코드

import itertools 
import pandas as pd

#이력서와 채용공고의 고유 ID리스트 추출
resume_ids=resume_df['resume_seq'].unique()
recruitment_ids=recruitment_df['recruitment_seq'].unique()

#이력서 아이디와 채용공고 아이디 의 모든 조합 생성
all_combinations=pd.DataFrame(list(itertools.product(resume_ids,
                                                     recruitment_ids)),
                             columns=['resume_seq','recruitment'])
#생성된 조합 확인
all_combinations.head()
```

```python
#이력서 데이터와 채용공고 데이터의 특성을 결합하고, 필요한 추가특성생성

#교육 특성을 기준으로 결합을 시도함

resume_features=resume_df[['resume_seq','degree']]

#채용 공고 데이터에서 필요한 특성 선택
recruitment_features=recruitment_df[['recruitment_seq','education']]

#이력서와 채용공고 데이터 결합
combined_features=pd.merge(
apply_train_df,
resume_features,
    on='resume_seq',
    how='left'
).merge(recruitment_features,
       on='recruitment_seq',
       how='left'
       )
```

```python
#추가 특성 생성 : 이력서의 학위와 채용공고의 요구 수준 비교
combined_features['education_match']=(combined_features['degree']==combined_features['education']).astype(int)
```
